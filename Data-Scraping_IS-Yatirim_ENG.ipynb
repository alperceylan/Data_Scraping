```python
import requests
from bs4 import BeautifulSoup
import time
import re
```

```python
class Stock:


    def __init__(self):
        self.loop = True



    def program(self):
        choice = self.menu()

        if choice == '1':
            print("Fetching Current Prices...\n")
            time.sleep(2)
            self.currentPrice()

        if choice == '2':
            print("Fetching Company Profile...\n")
            time.sleep(2)
            self.profile()

        if choice == '3':
            print("Fetching Current Values...\n")
            time.sleep(2)
            self.currentValues()

        if choice == '4':
            print("Fetching Yield Figures...\n")
            time.sleep(2)
            self.yieldFigures()

        if choice == '5':
            print("Fetching Index Weight Ratios...\n")
            time.sleep(2)
            self.IWR()

        if choice == '6':
            print("Exiting from Automation... Thank you.")
            time.sleep(2)
            self.exit()



    def menu(self):
        def control(select):

            if not re.search("[1-6]" , choice):
                raise Exception("Please make a valid choice between 1 and 6...")
            elif len(choice) != 1:
                raise Exception("Please make a valid choice between 1 and 6...")

        while True:
            try:
                choice=input("Hello, Welcome to the Stock Automation System... \n\n Please Select the Operation You Would Like to Perform... \n\n[1] - Current Price \n[2] - Company Profile \n[3] - Current Values \n[4] - Yield Figures \n[5] - Indices the Company is Included In \n[6] - Exit \n\n")
                control(choice)

            except Exception as error:
                print(error)
                time.sleep(2)
                
            else:
                break

        return choice



    def currentPrice(self):
        while True:
            try:
                company = input("Please Enter the 'Stock Code' : ")
                url = "https://www.isyatirim.com.tr/tr-tr/analiz/hisse/Sayfalar/default.aspx"\
                                            .format(company)

                parser = BeautifulSoup(requests.get(url).content , "html.parser")
                price = parser.find("a" , {'href' : '/tr-tr/analiz/hisse/Sayfalar/sirket-karti.aspx?hisse={}'\
                                            .format(company.upper())}).parent.parent.find_all('td')

                info1 = price[1].string
                info2 = price[2].span.string
                info3 = price[3].string
                info4 = price[4].string
                info5 = price[5].string

                print(f"Last Price : {info1} \n Change(%) : {info2.lstrip()} \r Change(TL) : {info3} \n Volume(TL) : {info4} \n Volume(Count) : {info5}")
                break

            except AttributeError:
                print('You Entered an Invalid Stock Code...')
                time.sleep(2)    
        time.sleep(2)
        self.returnMenu()



    def profile(self):
        while True:
            try:
                company = input("Please Enter the 'Stock Code' : ")
                url = "https://www.isyatirim.com.tr/tr-tr/analiz/hisse/Sayfalar/sirket-karti.aspx?hisse={}"\
                                            .format(company)
                parser = BeautifulSoup(requests.get(url).content , "html.parser")

                profile = parser.find("div" , {'id' : 'ctl00_ctl58_g_6618a196_7edb_4964_a018_a88cc6875488'})\
                                            .find_all("tr")
                for i in profile:
                    info1 = i.th.string
                    info2 = i.td.string
                    print(f'{info1}:{info2}')
                break

            except AttributeError:
                print('You Entered an Invalid Stock Code...')
                time.sleep(2)    
        time.sleep(2)
        self.returnMenu()



    def currentValues(self):
        while True:
            try:
                company = input("Please Enter the 'Stock Code' : ")
                url = "https://www.isyatirim.com.tr/tr-tr/analiz/hisse/Sayfalar/sirket-karti.aspx?hisse={}"\
                                            .format(company)
                parser = BeautifulSoup(requests.get(url).content , "html.parser")

                currentValues = parser.find("div" , {'id' : 'ctl00_ctl58_g_76ae4504_9743_4791_98df_dce2ca95cc0d'})\
                                            .find_all("tr")
                for i in currentValues:
                    info1 = i.th.string
                    info2 = i.td.string
                    print(f'{info1} : {info2}')
                break

            except AttributeError:
                print('You Entered an Invalid Stock Code...')
                time.sleep(2)    
        time.sleep(2)
        self.returnMenu()



    def yieldFigures(self):
        while True:
            try:
                company = input("Please Enter the 'Stock Code' : ")
                url = "https://www.isyatirim.com.tr/tr-tr/analiz/hisse/Sayfalar/sirket-karti.aspx?hisse={}"\
                                            .format(company)
                parser = BeautifulSoup(requests.get(url).content , "html.parser")

                yieldFigures = parser.find("div" , {'id' : 'ctl00_ctl58_g_aa8fd74f_f3b0_41b2_9767_ea6f3a837982'})\
                                            .find('table').find('tbody').find_all('tr')
                for i in yieldFigures:
                    info = i.find_all('td')
                    print(f'Unit: {info[0].string}   Daily(%): {info[1].string}   Weekly(%): {info[2].string}   Monthly(%): {info[3].string}   Year-to-Date Change(%): {info[4].string}')
                break

            except AttributeError:
                print('You Entered an Invalid Stock Code...')
                time.sleep(2)    
        time.sleep(2)
        self.returnMenu()



    def IWR(self):
        while True:
            try:
                company = input("Please Enter the 'Stock Code' : ")
                url = "https://www.isyatirim.com.tr/tr-tr/analiz/hisse/Sayfalar/sirket-karti.aspx?hisse={}"\
                                            .format(company)
                parser = BeautifulSoup(requests.get(url).content , "html.parser")

                IWR_1 = parser.find("div" , {'id' : 'ctl00_ctl58_g_655a851d_3b9f_45b0_a2d4_b287d18715c9'})\
                                            .find('table').find('tbody').find('tr').find_all('td')
                IWR_2 = parser.find("div" , {'id' : 'ctl00_ctl58_g_655a851d_3b9f_45b0_a2d4_b287d18715c9'})\
                                            .find('table').find('thead').find('tr').find_all('th')
                for i in range(0,3):
                    print(f"{IWR_2[i].string} : {IWR_1[i].string}")
                break

            except AttributeError:
                print('You Entered an Invalid Stock Code...')
                time.sleep(2)    
        time.sleep(2)
        self.returnMenu()



    def exit(self):
        self.loop = False
        exit()



    def returnMenu(self):
        while True:
            x = input("Press '0' to Return to the Main Menu... ")
            if x == '0':
                print('Returning to the Main Menu...')
                time.sleep(2)
                self.program()
                break
            else:
                print('Please Make a Valid Selection...')



system = Stock()
while system.loop:
    system.program()
```
